```python
#Regula perceptronului exemplu curs
#doar pentru fc de activare binara
import numpy as np
x1 = [1,-2,0,-1] #vect de intrare
x2 = [0,1.5,-0.5,-1]
x3 = [-1,1,0.5,-1]
w = [1,-1,0,0.5] #pondere
Rx1 = -1 #raspunsuri dorite 
Rx2 =-1
Rx3 = 1

def sgn(x):
    if x > 0 :
        sgn = 1
    elif x < 0:
        sgn = -1
    return sgn

def out(w,x):
    net = np.dot(x, w) #prod scalar
    print("net = ", net)
    sign = sgn(net)
    
    return sign



def f(x,w_prev,net): #refacere pondere
    #w = w_prev[j] + c *2 dc sgn=net = -1
    #w = w_prev[j] - c *2 dc sgn=net = 1
    c = 0.1
    w = []
    if net == -1:
        for j in range(len(x)):
            w.append(2 * c * x[j] + w_prev[j])
    elif net == 1:
        for j in range(len(x)):
            w.append(-2 *c * x[j] + w_prev[j])
    return w 



w_current = w 
i = 1
while i<2 :
    print("Iteratia: ",i)
    w_prev = w_current
    print("\nw_start este",w_prev)
    
    print("x1 este" ,x1)
    net = out(w_current,x1) 
    if Rx1 != net:
        w_current = f(x1,w_prev,net)
    print("w_next este:",w_current)
    
    w_prev = w_current
    print("\nw_start este",w_prev)
    
    print("x2 este" ,x2)
    net = out(w_current,x2)
    if Rx2 != net:
        w_current = f(x2,w_prev,net)
    print("w_next este:",w_current)
    
    
    w_prev = w_current
    print("\nw_start este",w_prev)
    
    print("x3 este" ,x3)
    net = out(w_current,x3)
    if Rx3 != net:
        w_current = f(x3,w_prev,net)
    print("w_next este:",w_current)
    
    print("\n\n")
    i=i+1
   
    
```
