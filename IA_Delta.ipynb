```python
#Regula delta
import numpy as np
import math 

y1 = [45,85,-1]
y2 = [50,43,-1]
y3 = [40,80,-1]
y4 = [55,42,-1]
y5 = [200,43,-1]
y6 = [48,40,-1]
y7 = [195,41,-1]
y8 = [43,87,-1]
y9 = [190,40,-1]

d1 = [1,-1,-1]
d2 = [-1,1,-1]
d3 = [1,-1,-1]
d4 = [-1,1,-1]
d5 = [-1,-1,1]
d6 = [-1,1,-1]
d7 = [-1,-1,1]
d8 = [1,-1,-1]
d9 = [-1,-1,1]

d = [d1,d2,d3,d4,d5,d6,d7,d8,d9]
w = [[1,-1,1],[-1,0.5,-0.5],[0.75,-0.2,0.5]]


#scalarea (normalizarea) pe coloane pentru a aduce val in int [0,1]:
list = [[45,85,-1],[50,43,-1],[40,80,-1],[55,42,-1],[200,43,-1],[48,40,-1],[195,41,-1],[43,87,-1],[190,40,-1]]

mylist = [y1,y2,y3,y4,y5,y6,y7,y8,y9]

npArray = np.array(mylist)

print('Pentru coloana 0:')
col0_min = np.min(npArray[:,0])
print('Valoarea minima de pe coloana 0 este', col0_min)
col0_max = np.max(npArray[:,0])
print('Valoarea maxima de pe coloana 0 este:', col0_max)

print('Pentru coloana 1:')
col1_min = np.min(npArray[:,1])
print('Valoarea minima de pe coloana 1 este', col1_min)
col1_max = np.max(npArray[:,1])
print('Valoarea maxima de pe coloana 1 este:', col1_max)

col0_norm = (npArray[:,0]-min(npArray[:,0])) / (max(npArray[:,0]) - min(npArray[:,0]))
#print('\n',col0_norm)

col1_norm = (npArray[:,1]-min(npArray[:,1])) / (max(npArray[:,1]) - min(npArray[:,1]))
#print('\n',col1_norm)
print('\n\n')
#lista finala 
col0 = np.array(col0_norm)
col1 = np.array(col1_norm)
y = []
for i in range(9):
    x1 = col0[i]
    x2 = col1[i]
    x3 = -1
    x = [x1,x2,x3]
    y.append(x)

print("Lista de vectori:",y)


# f(net)->f(net)=2/(1+exp**-y*net)-1, y>0
def fout(w,x):
    y = 1 
    net = np.dot(x, w)
    k = math.exp(-y*net)
    o = 2/(1+k)-1
    #print("f(net) este:",o)
    return o

epoch = 0
w_current = w
c = 0.01
E = 0
E_current = E
Emax = 2
while epoch < 100:
    
    E_current = E
    for i in range(len(y)):
    ###calc out - iesirile fiecarui perceptron
        w_prev = w_current
        E_prev = E_current
        for j in range(len(w_prev)):
            if j == 0:
                x1 = fout(w_current[j],y[i])
                out1 = x1*(y[i][0]*w_current[j][0]+y[i][1]*w_current[j][1]+y[i][2]*w_current[j][2])
            elif j == 1:
                x2 = fout(w_current[j],y[i])
                out2 = x2*(y[i][0]*w_prev[j][0]+y[i][1]*w_current[j][1]+y[i][2]*w_current[j][2])
            elif j == 2:
                x3 = fout(w_current[j],y[i])
                out3 = x3*(y[i][0]*w_prev[j][0]+y[i][1]*w_current[j][1]+y[i][2]*w_current[j][2])
    
    ###ref ponderi
        o = [out1,out2,out3]
        for m in range(3):
            for n in range(3):
                w_current[m][n] = w_prev[m][n]+c*(d[i][n]-o[n])*(1-(o[n]**2))*y[i][n]
    

    
    #eroarea cumulata:
        E_current = E_prev + ((d[i][0]-o[0])**2) + ((d[i][1]-o[1])**2) + ((d[i][2]-o[2])**2)
        print("\nPonderea refacuta este: ",w_current)
        print("\nE_current este: ", E_current)
        epoch = epoch + 1
        print("Eroarea este:",E_current)
    if E_current<Emax:
        break
```
