```python
#Castigatorul ia tot
import numpy as np
import math
#pattern:
x1 = [45,85]
x2 = [50,43]
x3 = [40,80]
x4 = [55,42]
x5 = [200,43]
x6 = [48,40]
x7 = [195,41]
x8 = [43,87]
x9 = [190,40]

list = [x1,x2,x3,x4,x5,x6,x7,x8,x9]
#prototipuri:
w1 = [60,70]
w2 = [70,40]
w3 = [150,30]

epoch = 100

def dist(x,w):
    dist = math.sqrt((x[0]-w[0])**2 + (x[1]-w[1])**2)
    return dist

def pondere(w,x):
    c = 0.1
    w_current = []
    i = 1
    for i in range(2):
        w_current.append(w[i] + c *(x[i] - w[i]))
      
    #print(" Ponderea recalculata este", w_current)
    return w_current
    

j = 0
min = 0 
i = 1
while i < epoch:
    print("Pentru epoca:",i)
   
    for j in range(len(list)):
        print("Punctul este",list[j])
        d1 = dist(list[j],w1)
        d2 = dist(list[j],w2)
        d3 = dist(list[j],w3)
   
        if d1<d2 and d1<d3:
            min = d1
            wi = pondere(w1,list[j])
            print("Prototipul castigator este:", wi)
        elif d2<d1 and d2<d3:
            min = d2
            wi = pondere(w2,list[j])
            print("Prototipul castigator este:", wi)
        elif d3<d1 and d3<d2:
            min = d3
            wi = pondere(w3,list[j])
            print("Prototipul castigator este:", wi)
        if d1 == min:
            w1 = wi
        elif d2 == min:
            w2 = wi
        elif d3 == min:
            w3 = wi
        
    i=i+1
    print("\n")
```
