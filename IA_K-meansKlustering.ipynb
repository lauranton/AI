```python
#K-means clustering
import numpy as np
import math
import random

from itertools import combinations
k = 3 #nr de clustere

P1 = [45,85]
P2 = [50,43]
P3 = [40,80]
P4 = [55,42]
P5 = [200,43]
P6 = [48,40]
P7 = [195,41]
P8 = [43,87]
P9 = [190,40]

P = [[45,85],[50,43],[40,80],[55,42],[200,43],[48,40],[195,41],[43,87],[190,40]]


def dist(x,y):
    dist = math.sqrt((x[0]-y[0])**2 + (x[1]-y[1])**2)
    return dist


def RefClus(cluster_list): #media pe coordonate
    x = []
    x = cluster_list 
    sumx = 0
    sumy = 0
    av1 = 0
    av2 = 0
    med1 = 0
    med2 = 0
    for i in range(len(x)):
        var = x[i] #copie a pc
        for j in range(len(var)):
            if j == 0:
                sumx = sumx + var[j]
                av1 = av1 + 1
            if j == 1:
                sumy = sumy + var[j]
                av2 = av2 + 1
    med1 = sumx / av1
    med2 = sumy / av2
    new_cluster = [med1,med2]
    return new_cluster
    

list = random.sample(P,3)

cluster1 = list[0]
cluster2 = list[1]
cluster3 = list[2]


cluster1_current = 0
cluster2_current = 0
cluster3_current = 0
i = 1

while i < 500:
    cluster1_list = []
    cluster2_list = []
    cluster3_list = []

    if i == 1:
        cluster1_current = cluster1
        cluster2_current = cluster2
        cluster3_current = cluster3
    
    print("Cluster curent:",cluster1_current,cluster2_current,cluster3_current)
    k = 1
    for j in range(len(P)):
        d1 = dist(cluster1_current,P[j])
        d2 = dist(cluster2_current,P[j])
        d3 = dist(cluster3_current,P[j])
        
        
        if d1<d2 and d1<d3:
            min = d1
        elif d2<d1 and d2<d3:
            min = d2
        elif d3<d1 and d3<d2:
            min = d3
        

        k = P[j] 
      
        if min == d1:
            cluster1_list.append(P[j])
            c1_x = k[0]
            c1_y = k[1]
        elif min == d2:
            cluster2_list.append(P[j])
            c2_x = k[0]
            c2_y = k[1]
        elif min == d3:
            cluster3_list.append(P[j])
            c3_x = k[0]
            c3_y = k[1]
            
        #print("\n")

    print("Pc atasate lista Cluster1 = ", cluster1_list)
    print("Pc atasate lista Cluster2 = ", cluster2_list)
    print("Pc atasate lista Cluster3 = ", cluster3_list)   
    
    #copie pt clustere pentru a vedea cand se opreste programul
    c1 = cluster1_current
    c2 = cluster2_current 
    c3 = cluster3_current
    
    
    cluster1_current = RefClus(cluster1_list)
    print("Cluster1 urmator este:",cluster1_current)
    cluster2_current = RefClus(cluster2_list)
    print("Cluster2 urmator este:",cluster2_current)
    cluster3_current = RefClus(cluster3_list)
    print("Cluster3 urmator este:",cluster3_current)
    print("\n")
   
    #verificare oprire porgram(dc si dupa refacerea clusterelor acestea raman la fel programul se opreste)
    if c1 == cluster1_current and c2 == cluster2_current and c3 == cluster3_current:
        break

    i = i+1
    
    
```
