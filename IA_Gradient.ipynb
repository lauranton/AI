```python
#f(x)=6*x**2 - 12*x +1

current_x = 3 #inceputul algoritmului
C = 0.01 #constanta de invatare
diff = 0.00001 #finalul algoritmului-> cand modificarile componentelor x scad sub aceasta valoare
prev_step_size = 1
max_i = 10000 #nr maxim de pasi
i = 0
df = lambda x : 12 * x - 12 #derivata fuctiei 


while (prev_step_size > diff and i < max_i):
    prev_x = current_x
    current_x = current_x - C * df(prev_x)
    prev_step_size = abs(current_x - prev_x)
    i = i + 1
    print("Iteratia", i,":", "\nValoarea lui x este: ",current_x)

print("Minimul este: ", current_x)


```

```python
#g(x,y) = x**2 + 2 * y**2

current_x = 3 #inceputul algoritmului
current_y = 3
C = 0.01
diff = 0.00001 #finalul algoritmului-> cand modificarile componentelor x scad sub aceasta valoare
prev_step_sizeX = 1
prev_step_sizeY = 1
max_i = 10000 #nr maxim de pasi
i = 0
dfx = lambda x : 2*x #derivata fuctiei in fc de x
dfy = lambda y : 4*y #derivata functiei in fc de y

while (prev_step_sizeX > diff and prev_step_sizeY > diff and i < max_i):
    prev_x = current_x
    prev_y = current_y
    current_x = current_x - C * dfx(prev_x)
    current_y = current_y - C * dfy(prev_y)
    prev_step_sizeX = abs(current_x - prev_x)
    prev_step_sizeY = abs(current_y - prev_y)
    i = i + 1
    print("Iteratia", i,":", "\nValoarea lui x: ",current_x, "valoarea lui y: ", current_y)

print("Minimul este: ", current_x, current_y)




```
