```python
#invatare hebbiana ex curs pg 43
#1 neuroni bipolari binari f(net)=sgn(net)
import numpy as np

w1 = [ 1, -1, 0, 0.5] #pondere initiala
x1 = [1, -2, 1.5, 0]
x2 = [1, -0.5, -2, -1.5]
x3 = [0, 1, -1, 1.5]

c = 1

def sgn(x):
    if x > 0 :
        sgn = 1
    elif x < 0:
        sgn = -1
    return sgn

def out(w,x): # sgn prod scalar
    net = np.dot(x, w) 
    sign = sgn(net)
    print("f(net) este:",sign)
    return sign

def f(x,w_prev,net): #pondere noua
    w = []
    for j in range(len(x)):
        w.append(net * x[j] + w_prev[j])
    return w 
  
w_current = w1
i = 1
while i < 6 :
    print("Iteratia: ",i)
    w_prev = w_current
    print("w_start este",w_prev)
   
    print("\nx1 este" ,x1)
    net = out(w_current,x1)
    w_current = f(x1,w_prev,net)
    print("w_next este:",w_current)
    
    w_prev = w_current
    print("\nx2 este",x2)
    net = out(w_current,x2)
    w_current = f(x2,w_prev,net)
    print("w_next este:",w_current)
    
    
    w_prev = w_current
    print("\nx3 este",x3)
    net = out(w_current,x3)
    w_current = f(x3,w_prev,net)
    print("w_next este:",w_current)
    print("\n\n")
    i=i+1
   
    

```
