```python
#Regresia liniara
import csv
import numpy as np
import random
myList = []
with open(r'C:\Users\laura\Salary_Data.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
             if len(row) == 2:
                myList.append([float(row[0]), float(row[1])])
            
print(myList)


prev_w1 = random.randint(1,50) #inceputul algoritmului
prev_w2 = random.randint(1,50) #inceputul algoritmului
c = 0.01
diff = 0.00001 #finalul algoritmului-> cand modificarile componentelor x scad sub aceasta valoare

n = 15

while(True):
    s1 = 0
    s2 = 0
    for i in range(15): #calcul derivate erori
        s1 =s1+( myList[i][1] - (prev_w1 * myList[i][0] + prev_w2)) * myList[i][0]
        s2 = s2+ myList[i][1] - (prev_w1 * myList[i][0] + prev_w2)     
    
    s1 = -s1/n
    s2 = -s2/n
    current_w1 = prev_w1-c*s1
    current_w2 = prev_w2-c*s2
    
    o1 = abs(current_w1 - prev_w1)
    o2 = abs(current_w2 - prev_w2)
    print("w1 este:", current_w1)
    print("w2 este:", current_w2)
 
    if o1 <= diff  and o2 <= diff:
        break
    else:
        prev_w1 = current_w1
        prev_w2 = current_w2
```
